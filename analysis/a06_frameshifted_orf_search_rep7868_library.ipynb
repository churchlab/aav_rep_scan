```python
import Bio
from Bio.Data import CodonTable
from statistics import mean
import os 
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import seaborn as sns
import math
```

```python
selection_value_dir = '../analysis/selection_values/'
output_dir = '../analysis/output_figures/'
```

```python
rep7868_selection_codon_df = pd.read_csv(selection_value_dir + 'rep7868_selection_values_codon_trans.csv',
                                         header=[0], index_col=[0,1,2,3,4,5], skipinitialspace=True)
```

```python
standard_table = CodonTable.unambiguous_dna_by_name["Standard"]
codon_dict = standard_table.forward_table
# Add stop codons
codon_dict['TAA'] = '*'
codon_dict['TAG'] = '*'
codon_dict['TGA'] = '*'
codons = codon_dict.keys()
# codon_dict
```

```python
# Read in df with oligo designs
complete_rep_oligos_df = pd.read_csv('../library_design/20210325_aav2_rep_complete_oligo_seqs_tiles0-13.csv', index_col=0)
rep7868_lib_tiles = [0, 1, 2, 11, 12, 5, 6, 7, 8, 9, 10]
rep7868_oligos_df = complete_rep_oligos_df.loc[complete_rep_oligos_df['tile_num'].isin(rep7868_lib_tiles)]
print (len(wt_rep_oligos_df.index))
wt_rep_oligos_df.head()
```

```python
# For each BC in wt_rep_oligos determine what the codon in the +1 and +2 frames are, add new columns with this info
wt_rep_oligos_dict = {}
for o, oligo in enumerate(wt_rep_oligos_df.iterrows()):
#     if o > 1000: break
    aa = oligo[1][0]
    abs_pos = oligo[1][1]
    codon = oligo[1][4]
    barcode = oligo[1][2]
    is_wt = oligo[1][7]
    is_wt_aa = oligo[1][8]
    lib_type = oligo[1][9]
    tile_num = oligo[1][15]
    mutant_check = oligo[1][10]
    new_swap = oligo[1][12]
#     print (mutant_check, new_swap)
    next_codon = new_swap[9:12]
#     print (next_codon)
    frame2_codon = codon[1:3] + next_codon[0]
    frame3_codon = codon[2] + next_codon[0:2]
    frame2_aa = codon_dict[frame2_codon]
    frame3_aa = codon_dict[frame3_codon]
#     print (frame2_aa)
#     print (frame3_aa)
    wt_rep_oligos_dict[barcode] = [aa, abs_pos, codon, is_wt, is_wt_aa, lib_type, tile_num, frame2_codon, frame3_codon,
                                  frame2_aa, frame3_aa]
wt_rep_oligos_with_frameshift_info = pd.DataFrame.from_dict(wt_rep_oligos_dict, columns = ['aa', 'abs_pos', 'codon',
                                                                                           'is_wt', 'is_wt_aa', 'lib_type',
                                                                                           'tile_num', 'frame2_codon',
                                                                                           'frame3_codon', 'frame2_aa', 
                                                                                           'frame3_aa'], orient='index')
# wt_rep_oligos_with_frameshift_info.to_csv('../output_csv/rep7868_library_oligo_design_with_frameshift_info.csv')
wt_rep_oligos_with_frameshift_info.head()
```

```python
wt_rep_oligos_with_frameshift_info.drop_duplicates(subset=['abs_pos', 'codon', 'lib_type'], inplace=True)
wt_rep_oligos_with_frameshift_info.reset_index(inplace=True)
wt_rep_oligos_with_frameshift_info.drop('index', axis=1, inplace=True)
wt_rep_oligos_with_frameshift_info.head()
```

```python
rep7868_selection_codon_df.reset_index(inplace=True)
rep7868_selection_codon_df.head()
```

```python
codon_level_selection_with_frameshift_info = rep7868_selection_codon_df.merge(wt_rep_oligos_with_frameshift_info, how='inner', on=['abs_pos', 'aa', 'codon', 'lib_type'])
print (len(codon_level_selection_with_frameshift_info.index))
codon_level_selection_with_frameshift_info.head()
```

```python
codon_level_selection_with_frameshift_info.drop(['is_wt_x'], axis=1, inplace=True)
codon_level_selection_with_frameshift_info.columns = codon_level_selection_with_frameshift_info.columns.str.replace('is_wt_y', 'is_wt')
codon_level_selection_with_frameshift_info.set_index(['abs_pos', 'aa', 'codon', 'aa-codon', 'lib_type', 'is_wt', 'is_wt_aa',
                                                     'tile_num', 'frame2_codon', 'frame3_codon', 'frame2_aa', 'frame3_aa'],
                                                     inplace=True)
codon_level_selection_with_frameshift_info.head()
```

```python
# Create dfs with selection values just for variants with stop codons in frame 2 or frame 3
selection_frame2_stops = codon_level_selection_with_frameshift_info.xs('*', level='frame2_aa', drop_level=False)
selection_frame3_stops = codon_level_selection_with_frameshift_info.xs('*', level='frame3_aa', drop_level=False)
selection_frame3_stops.head()
```

```python
# Get list of unique abs_pos and aa where there is a stop codon in frame 2
frame2_stops_index = np.array(selection_frame2_stops.index.tolist())
frame2_stops_aa_pos = frame2_stops_index[:, 0:2]
frame2_stops_unique_aa_pos = np.vstack({tuple(row) for row in frame2_stops_aa_pos})
```

```python
# Subset codon_level_selection_with_frameshift_info to include only abs_pos and aa that have a stop in frame2
# Will use this df as control for frameshifted ORF analysis
synonymous_to_frame2_stops = pd.DataFrame(columns=['abs_pos', 'aa', 'codon', 'aa-codon', 'lib_type', 'is_wt', 'is_wt_aa',
                                                     'tile_num', 'frame2_codon', 'frame3_codon', 'frame2_aa', 'frame3_aa', '1'])

synonymous_to_frame2_stops.set_index(['abs_pos', 'aa', 'codon', 'aa-codon', 'lib_type', 'is_wt', 'is_wt_aa',
                                      'tile_num', 'frame2_codon', 'frame3_codon', 'frame2_aa', 'frame3_aa'],
                                     inplace=True)

for r, row in frame2_stops_unique_aa_pos:
    selection_data_all_synonymous = codon_level_selection_with_frameshift_info.xs((int(r), row), axis=0, level=['abs_pos', 'aa'],
                                                                                 drop_level=False)
    synonymous_to_frame2_stops = pd.concat([synonymous_to_frame2_stops, selection_data_all_synonymous])
synonymous_to_frame2_stops.head()
```

```python
def save_fig(fig, fig_name, directory, transparent=True):
    fig_path = os.path.join(directory, fig_name)
    fig.savefig(fig_path, bbox_inches='tight', dpi='figure',transparent=transparent)
```

```python
selection_frame2_stops_avg_by_pos = selection_frame2_stops.groupby(level=['abs_pos']).mean()
synonymous_to_frame2_stops_avg_by_pos = synonymous_to_frame2_stops.groupby(level=['abs_pos']).mean()

fig,ax = plt.subplots(figsize=[6,5])
ax.set_xlabel('codon position', fontsize=16)
ax.set_ylabel("log2(s')", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim((-2.5, 1))
plt.xlim((0, 622))
ax.scatter(list(selection_frame2_stops_avg_by_pos.reset_index()['abs_pos']), selection_frame2_stops_avg_by_pos['virus'].apply(np.log2),s=.5, c='indianred')
ax.plot(list(selection_frame2_stops_avg_by_pos.reset_index()['abs_pos']), selection_frame2_stops_avg_by_pos['virus'].rolling(10,min_periods=1).mean().apply(np.log2),c='indianred', lw=1)

ax.scatter(list(synonymous_to_frame2_stops_avg_by_pos.reset_index()['abs_pos']), synonymous_to_frame2_stops_avg_by_pos['virus'].apply(np.log2),s=.5, c='black')
ax.plot(list(synonymous_to_frame2_stops_avg_by_pos.reset_index()['abs_pos']), synonymous_to_frame2_stops_avg_by_pos['virus'].rolling(10,min_periods=1).mean().apply(np.log2),c='black', lw=1)
plt.legend(frameon=False)
save_fig(fig, 'rep7868_frameshifted_ORF_analysis_plus_one.png', output_dir)
```

```python
# Get list of unique abs_pos and aa where there is a stop codon in frame 2
frame3_stops_index = np.array(selection_frame3_stops.index.tolist())
frame3_stops_aa_pos = frame3_stops_index[:, 0:2]
frame3_stops_unique_aa_pos = np.vstack({tuple(row) for row in frame3_stops_aa_pos})
# frame3_stops_unique_aa_pos = np.unique(frame3_stops_aa_pos_new)
frame3_stops_unique_aa_pos

# Subset codon_level_selection_with_frameshift_info to include only abs_pos and aa that have a stop in frame3
# Will use this df as control for frameshifted ORF analysis
synonymous_to_frame3_stops = pd.DataFrame(columns=['abs_pos', 'aa', 'codon', 'aa-codon', 'lib_type', 'is_wt', 'is_wt_aa',
                                                     'tile_num', 'frame2_codon', 'frame3_codon', 'frame2_aa', 'frame3_aa', '1'])

synonymous_to_frame3_stops.set_index(['abs_pos', 'aa', 'codon', 'aa-codon', 'lib_type', 'is_wt', 'is_wt_aa',
                                      'tile_num', 'frame2_codon', 'frame3_codon', 'frame2_aa', 'frame3_aa'],
                                     inplace=True)
# synonymous_to_frame3_stops

for r, row in frame3_stops_unique_aa_pos:
    selection_data_all_synonymous = codon_level_selection_with_frameshift_info.xs((int(r), row), axis=0, level=['abs_pos', 'aa'],
                                                                                 drop_level=False)
    synonymous_to_frame3_stops = pd.concat([synonymous_to_frame3_stops,selection_data_all_synonymous])
synonymous_to_frame3_stops.head()
```

```python
selection_frame3_stops_avg_by_pos = selection_frame3_stops.groupby(level=['abs_pos']).mean()
synonymous_to_frame3_stops_avg_by_pos = synonymous_to_frame3_stops.groupby(level=['abs_pos']).mean()

fig,ax = plt.subplots(figsize=[6,5])
ax.set_xlabel('codon position', fontsize=16)
ax.set_ylabel("log2(s')", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim((-2.5, 1))
# plt.legend()
ax.scatter(list(selection_frame3_stops_avg_by_pos.reset_index()['abs_pos']), selection_frame3_stops_avg_by_pos['virus'].apply(np.log2),s=.5, c='indianred')
ax.plot(list(selection_frame3_stops_avg_by_pos.reset_index()['abs_pos']), selection_frame3_stops_avg_by_pos['virus'].rolling(10,min_periods=1).mean().apply(np.log2),c='indianred', lw=1)

ax.scatter(list(synonymous_to_frame3_stops_avg_by_pos.reset_index()['abs_pos']), synonymous_to_frame3_stops_avg_by_pos['virus'].apply(np.log2),s=.5, c='black')
ax.plot(list(synonymous_to_frame3_stops_avg_by_pos.reset_index()['abs_pos']), synonymous_to_frame3_stops_avg_by_pos['virus'].rolling(10,min_periods=1).mean().apply(np.log2),c='black', lw=1)
plt.legend(frameon=False)
save_fig(fig, 'rep7868_frameshifted_ORF_analysis_plus_two.png', output_dir)
```
